```python
import requests
import pandas as pd
import os
import certifi
import urllib3

# Disable SSL warnings
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# Load station IDs from CSV exported from QGIS
stations_df = pd.read_csv('C:\\Users\\zomav\\Documents\\Estaciones.csv')
station_ids = stations_df.iloc[:, 0].astype(str).tolist()  # Assumes first column has IDs

print(station_ids)     

# Output folder
output_folder = 'Climatic_Stations_Data'
os.makedirs(output_folder, exist_ok=True)

# Base URL
base_url = 'http://smn.conagua.gob.mx/tools/RESOURCES/Normales_Climatologicas/Mensuales/mich/mes{}.txt'

for station_id in station_ids:
    url = base_url.format(station_id)
    response = requests.get(url, verify=False)

    try:
        response = requests.get(url)
        response.raise_for_status()
        raw_text = response.text

        # Parse text into rows
        lines = raw_text.strip().split('\n')
        data = [line.split() for line in lines if line.strip()]

        # Create DataFrame
        df = pd.DataFrame(data)

        # Save to Excel
        output_path = os.path.join(output_folder, f'{station_id}.xlsx')
        df.to_excel(output_path, index=False, header=False)
        print(f'✅ Saved: {output_path}')
    except Exception as e:
        print(f'❌ Failed for station {station_id}: {e}')
```
